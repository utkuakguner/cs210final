```python
import pandas as pd  
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint
import operator
from scipy import stats
from os.path import join 
import json
import datetime
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
import warnings
from sklearn import tree
from sklearn.metrics import accuracy_score
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data_path = "./data"
filename1 = "drug_overdose.csv"
filename2 = "anti_drug_spending.csv"
        
csv1 = pd.read_csv(join(data_path, filename1), delimiter=",")
csv2 = pd.read_csv(join(data_path, filename2), delimiter=",")
```

```python
def truncate(n, decimals=0):
    multiplier = 10 ** decimals
    return int(n * multiplier) / multiplier
```

```python
file1 = json.loads(csv1.to_json())
file2 = json.loads(csv2.to_json())

df1 = pd.DataFrame(file1)
df2 = pd.DataFrame(file2)

ratios1 = []
ratios2 = []

for i in range(6):
    ratios1.append(truncate((df2.Budget[i + 1] / df2.Budget[i] - 1) * 100, 2))
```

```python
plt.subplots(figsize=(10,10))
plt.bar(df2.Year, df2.Budget, color='green', width=.6)
plt.xlabel("Year")
plt.ylabel("Spending on war on drugs(in million $)")
plt.title("Spendings on war on drugs form 2012 to 2018")

for i,j in zip(df2.Year,df2.Budget):
    plt.annotate(str(j),xy=(i - 0.275,j + 300))

plt.show()
```

```python
years = {}

dates = df1.Date

for i in dates:
    year = 0
    if type(i) is str:
        year = int(i[6:10])
    else:
        year = 2012
    
    if year in years:
        years[year] += 1
    else:
        years[year] = 1
        
for i in range(6):
    ratios2.append(truncate((years[i + 2013] / years[i + 2012] - 1) * 100, 2))
```

```python
plt.subplots(figsize=(10,10))
plt.xlabel("Year")
plt.ylabel("Overdose deaths caused by drugs")
plt.title("Overdose death caused by drugs form 2012 to 2018")

plt.bar(years.keys(), years.values(), color='red', width=.6)
for i,j in zip(years.keys(),years.values()):
    plt.annotate(str(j),xy=(i - 0.15,j + 8))

plt.show()
```

```python
regBudgetArray = []
regDeathsArray = []

for i in range(df2.Year.size):
    regBudgetArray.append([int(df2.Budget[i])])

for i in range(len(years)):
    regDeathsArray.append([int(years[2012 + i])])
```

```python
regression_model = LinearRegression()
regression_model.fit(regBudgetArray, regDeathsArray)
plt.subplots(figsize=(10,10))
plt.xlabel("Budget spendings")
plt.ylabel("Death count")
plt.title("Regresion model for budget spendings and death counts")

line_pred = np.dot(regBudgetArray, regression_model.coef_.transpose()) + regression_model.intercept_

plt.plot(regBudgetArray, regDeathsArray , 'go', regBudgetArray, line_pred)
axes = plt.gca()
plt.show()
```

```python
_, p_value1 = stats.ttest_ind(a=regBudgetArray, b=regDeathsArray, equal_var=False)

if (p_value1 > 0.025):
    print("US Budget on war on drugs has effect on death caused by drug overdose")
else:
    print("US Budget on war on drugs has NO effect on death caused by drug overdose")
print("p_value for estimation was:", p_value1, "and  was set to 0.025")
```

```python
plt.subplots(figsize=(10,10))
plt.xlabel("Year Intervals")
plt.ylabel("Change on data by percentage")
plt.title("Change on data per year by percentage")

y = ("2012-2013", "2013-2014", "2014-2015", "2015-2016", "2016-2017", "2017-2018")

plt.plot(y, ratios1, color="blue")
plt.plot(y, ratios2, color="red")

plt.show()
```

```python
regression_model = LinearRegression()
regression_model.fit(regRatio1Array, regRatio2Array)
plt.subplots(figsize=(10,10))
plt.xlabel("Change ratio of budget spendings")
plt.ylabel("Change ratio of death count")
plt.title("Regresion model for change ratios of budget spendings and death counts")

line_pred = np.dot(regRatio1Array, regression_model.coef_.transpose()) + regression_model.intercept_

plt.plot(regRatio1Array, regRatio2Array , 'go', regRatio1Array, line_pred)
axes = plt.gca()
plt.show()
```

```python
_, p_value1 = stats.ttest_ind(a=regRatio1Array, b=regRatio2Array, equal_var=False)

if (p_value1 > 0.025):
    print("US Budget on war on drugs has effect on death caused by drug overdose")
else:
    print("US Budget on war on drugs has NO effect on death caused by drug overdose")
print("p_value for estimation was:", p_value1, "and  was set to 0.025")
```
